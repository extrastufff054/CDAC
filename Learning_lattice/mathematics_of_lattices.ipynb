Generate a lattice 

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def generate_lattice(v1, v2, bounds) : 
    points = []
    for i in range (bounds[1]) :
        for j in range(bounds[1]) :
            point = i*v1 + j*v2
            points.append(point) 

    return np.array(points)
```

Define your lattice vectors and bounds

```python
v1 = np.array([1.2, 0.1])
v2 = np.array([0.2,1.1])

bounds  = (5,5)  # Number of points in each direction
```

```python
lattice_points = generate_lattice(v1, v2, bounds)
```

Plotting the lattice points 

```python
plt.scatter(lattice_points[:, 0], lattice_points[:, 1])
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.title('2D Lattice Points')
plt.grid()
plt.show()
```

Explanation :

* Vectors: v1 and v2 define the lattice structure.
* Bounds: The bounds parameter determines how many points to generate in each direction.
* Plotting: Matplotlib is used to visualize the generated lattice points.

### Step 2 : Create a Lattice 

```python
from lattpy import simple_square
import matplotlib.pyplot as plt

# Initialize a square lattice
latt = simple_square(a = 1.0, neighbors = 1)
latt.build(shape = (5,3))  # Build a lattice of shape 5x3

# Plotting the lattice 
latt.plot()
plt.title('Square lattice')
plt.show()
```
